```python
import turtle
pen = turtle.Turtle()
pen.pensize(5)
def draw_blocks(x,y):
    
    for i in range(2):
        pen.fillcolor("red")
        pen.begin_fill()
        pen.forward(90)
        pen.right(90)
        pen.forward(150)
        pen.right(90)
        pen.end_fill()
    
    pen.penup()
    pen.goto(x,y)
    pen.pendown()
    
    for i in range(2):
        pen.fillcolor("white")
        pen.begin_fill()
        pen.forward(30)
        pen.right(90)
        pen.forward(90)
        pen.right(90)
        pen.end_fill()
              
draw_blocks(30,-30)
pen.penup()
pen.goto(90,0)
pen.pendown()
draw_blocks(120,-30)
pen.penup()
pen.goto(0,-150)
pen.pendown()
draw_blocks(30,-180)
pen.penup()
pen.goto(90,-150)
pen.pendown()
draw_blocks(120,-180)
pen.penup()
pen.goto(180,-150)
pen.pendown()

for i in range(4):
    pen.fillcolor("blue")
    pen.begin_fill()
    pen.forward(150)
    pen.right(90)
    pen.end_fill()
    
pen.penup()
pen.goto(230,-225)
pen.pendown()
pen.fillcolor("white")
pen.begin_fill()
pen.forward(50)
pen.right(90)
pen.forward(75)
pen.right(90)
pen.forward(50)
pen.right(90)
pen.forward(75)
pen.end_fill()
```
